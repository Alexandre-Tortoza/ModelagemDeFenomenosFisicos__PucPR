Escreva uma função em Python que compute o valor da função abaixo, emitindo uma mensagem
de erro caso a função receba um valor não permitido. Além disso, a função deverá plotar um
gráfico da função no intervalo entre ±10, no domínio dos números reais.
$$
f(x) = \frac{1}{\sqrt{2-X}}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 1 / np.sqrt(2 - x)

x_values = np.linspace(-10, 2, 1000)
y_values = f(x_values)


while True:
    highlight_x = float(input("\n Escolha um valor para x: "))
    if highlight_x < 2:
        break
    else:
        print("O valor de x deve estar dentro do intervalo Permitido")

highlight_y = f(highlight_x)

plt.plot(x_values, y_values)
plt.scatter(highlight_x, highlight_y, color='red', label=f'Ponto de destaque ({highlight_x}, {highlight_y:.2f})')
plt.title(r'Gráfico de $f(x) = \frac{1}{\sqrt{2 - x}}$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axvline(x=2, color='r', linestyle='--', label='x = 2')
plt.legend()
plt.show()
```

Utilizando a linguagem Python, e somente as bibliotecas Matplotlib e Numpy, crie uma função
para plotar uma raquete de tênis de mesa. No seu gráfico a raquete deve ser plotada de forma
que seja possível observar tanto o cabo quanto a área de rebatimento

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_raquete():
    cabo_x = np.array([0.4, 0.6, 0.6, 0.4, 0.4])
    cabo_y = np.array([-0.2, -0.2, 0.28, 0.28, -0.2])

    raio = 0.4
    angulo = np.linspace(0, 2*np.pi, 100)
    area_rebate_x = raio * np.cos(angulo) + 0.5
    area_rebate_y = raio * np.sin(angulo) + 0.5

    plt.fill(cabo_x, cabo_y, color='#deb887')
    plt.fill(area_rebate_x, area_rebate_y, color='#7e0106')
    plt.gca().set_aspect('equal', adjustable='box')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('Raquete de Tênis de Mesa')
    plt.show()

plot_raquete()
```

Utilizando a linguagem Python, nenhuma biblioteca extra e as técnicas de List Comprehension
crie uma função que devolva o resultado da fórmula a seguir para 𝑛 elementos.
$$
f(x) = 4 \left(1 - \sum_{1}^{\
n} \frac{1}{1 + 2n}\right)
$$

```python
def calcular_formula(n):
    soma = sum([1 / (1 + 2*i) for i in range(1, n+1)])
    resultado = 4 * (1 - soma)
    return resultado

def main():
    n = int(input('Qual o valor de N: '))
    resultado = calcular_formula(n)
    print("O resultado da fórmula é:", resultado)

if __name__ == "__main__":
    main()
```

Usando a linguagem Python, sem utilizar as bibliotecas Sympy e Numpy, escreva uma função
que receba um valor 𝑛 no domínio dos números reais e devolva o limite da função a seguir
quando 𝑥 tende a 𝑛

$$
f(x) = \frac{3x^2-1}{\sqrt{x}}
$$

```python
def limite_funcao(n):
    limite = (3*n**2 - 1) / (n**0.5)
    return limite

# Exemplo de uso:
n = float(input("valor de n: "))
limite = limite_funcao(n)
print("O limite da função quando x tende a", n, "é:", limite)
```
