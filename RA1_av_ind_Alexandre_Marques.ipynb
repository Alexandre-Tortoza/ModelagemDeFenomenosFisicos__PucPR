**Resolução**

01 -

Serie de taylor

$$\cos(x) = 1 - \frac{x^2}{2!} + \frac{x^4}{4!} - \frac{x^6}{6!} + \cdots$$

Substituindo isso na expressão original, temos:

$$\lim_{{x \to 0}} \frac{1 - (1 - \frac{x^2}{2!} + \frac{x^4}{4!} - \frac{x^6}{6!} + \cdots)}{x^2}$$

Simplificando:

$$\lim_{{x \to 0}} \frac{\frac{x^2}{2!} - \frac{x^4}{4!} + \frac{x^6}{6!} - \cdots}{x^2}$$

Dividindo os termos por x² :

$$\lim_{{x \to 0}} \frac{1}{2!} - \frac{x^2}{4!} + \frac{x^4}{6!} - \cdots$$

Como tende a 0, vão para zero. Portanto, o limite é:

$$\frac{1}{2}$$


```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, cos, pi, N

# Defina a variável simbólica x
x = symbols('x')

# Defina a função
f = (1 - cos(x)) / x**2

# Crie um array de valores de x
x_vals = np.linspace(-1, 1, 1000)

# Avalie a função nos pontos x_vals
y_vals = [N(f.subs(x, val)) for val in x_vals]

# Plote o gráfico
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label=r"$\frac{1 - \cos(x)}{x^2}$", color='b')
plt.axhline(y=1/2, color='r', linestyle='--', label=r"$y = \frac{1}{2}$")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Limite: $\\lim_{{x \\to 0}} \\frac{{1 - \\cos(x)}}{{x^2}}$')
plt.grid(True)
plt.legend()
plt.show()
```

**Resolução 02:**

x² - 1  pode ser fatorada para  (x+1) * (x-1). Multiplicamos o numerador e o denominador pelo conjugado do numerador para eliminar a raiz quadrada:

$$
\lim_{{x \to 1}} \frac{\sqrt{x + 8} - 3}{x^2 - 1} = \lim_{{x \to 1}} \frac{(\sqrt{x + 8} - 3)(\sqrt{x + 8} + 3)}{(x^2 - 1)(\sqrt{x + 8} + 3)}
$$

Aplicando a diferença de quadrados no numerador:

$$
= \lim_{{x \to 1}} \frac{x + 8 - 9}{(x+1)(x-1)(\sqrt{x + 8} + 3)}
$$

Simplificando o numerador:

$$
= \lim_{{x \to 1}} \frac{x - 1}{(x+1)(x-1)(\sqrt{x + 8} + 3)}
$$

Cancelando x - 1 no numerador e denominador:

$$
= \lim_{{x \to 1}} \frac{1}{(x+1)(\sqrt{x + 8} + 3)}
$$

Substituindo x por 1:

$$
= \frac{1}{(1+1)(\sqrt{1 + 8} + 3)} = \frac{1}{2 \cdot 6} = \frac{1}{12}
$$

Portanto, o limite é de fato $$\frac{1}{12} $$


```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *

# Define a função
def f(x):
    return (np.sqrt(x + 8) - 3) / (x**2 - 1)

# Cria um array de valores x de 0 a 2
x = np.linspace(0, 2, 400)
x = np.delete(x, np.where(x==1))  # Remove x=1 para evitar divisão por zero

# Cria um array de valores y usando a função
y = f(x)

# Cria uma figura e um conjunto de subplots
fig, ax = plt.subplots()

# Plota a função
ax.plot(x, y, label=r'$\frac{\sqrt{x + 8} - 3}{x^2 - 1}$')

# Adiciona um título
ax.set_title('Gráfico da função')

# Adiciona rótulos x e y
ax.set_xlabel('x')
ax.set_ylabel('f(x)')

# Adiciona uma legenda
ax.legend()

# Calcula o limite
limit_value = 1/12

# Plota o limite
ax.plot(1, limit_value, 'ro')  # 'ro' significa círculo vermelho
ax.annotate('Limite (1, 1/12)', (1, limit_value), textcoords="offset points", xytext=(-10,-10), ha='center')

# Exibe o gráfico
plt.show()
```

**Resolução 03:**
Podemos expandir o termo (2 + h)³ e simplificar:

$$
\lim_{h \to 0} \frac{(2 + h)^3 - 8}{h}
$$

Expandindo (2 + h)³, temos:

$$
\lim_{h \to 0} \frac{(2^3 + 3 \cdot 2^2 \cdot h + 3 \cdot 2 \cdot h^2 + h^3) - 8}{h}
$$

Simplificando 2³ e -8, obtemos:

$$
\lim_{h \to 0} \frac{(8 + 12h + 6h^2 + h^3) - 8}{h}
$$

Cancelando os termos +8 e -8:

$$
\lim_{h \to 0} \frac{12h + 6h^2 + h^3}{h}
$$

Agora, podemos simplificar h em todos os termos do numerador:

$$
\lim_{h \to 0} (12 + 6h + h^2)
$$

Como h se aproxima de zero, os termos 6h e h^2 também se aproximam de zero. Portanto, o limite é:

$$
\lim_{h \to 0} (12 + 6h + h^2) = 12
$$

Então, o resultado do limite é **12**.

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, limit

# Define a função para o limite
def f(h):
    return (2 + h)**3 - 8

# Define a variável e calcula o limite usando sympy
h = symbols('h')
lim = limit((2 + h)**3 - 8, h, 0)

# Cria um array de valores de h em torno de 0
h_values = np.linspace(-0.1, 0.1, 400)
# Calcula os valores da função sem o denominador (já que estamos olhando para o limite quando h se aproxima de 0)
function_values = [(f(val) - 8) / val if val != 0 else lim for val in h_values]

# Plota o gráfico
plt.plot(h_values, function_values, label='f(h)')
plt.axhline(y=lim, color='r', linestyle='--', label=f'Limite = {lim}')
plt.title('Gráfico mostrando o limite')
plt.xlabel('h')
plt.ylabel('f(h)')
plt.legend()
plt.grid(True)
plt.show()
```

Racionalizamos o numerador multiplicando e dividindo pela conjugada do numerador:

$$
= \lim_{t \to 0} \left( \frac{(\sqrt{t + 1} - 1)(\sqrt{t + 1} + 1)}{t\sqrt{t + 1}(\sqrt{t + 1} + 1)} \right)
$$

Simplificando o numerador:

$$
= \lim_{t \to 0} \left( \frac{t + 1 - 1}{t\sqrt{t + 1}(\sqrt{t + 1} + 1)} \right)
$$

$$
= \lim_{t \to 0} \left( \frac{t}{t\sqrt{t + 1}(\sqrt{t + 1} + 1)} \right)
$$

Cancelamos o 't' no numerador e no denominador:

$$
= \lim_{t \to 0} \left( \frac{1}{\sqrt{t + 1}(\sqrt{t + 1} + 1)} \right)
$$

Agora, aplicamos o limite quando t  tende a zero:

$$
= \frac{1}{\sqrt{0 + 1}(\sqrt{0 + 1} + 1)}

$$

$$
