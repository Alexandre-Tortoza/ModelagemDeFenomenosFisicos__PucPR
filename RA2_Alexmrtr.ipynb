# Questão 3 - Vale até 4 Pontos

Usando apenas ferramentas computacionais e gráficos, faça a análise dos pontos críticos das seguintes funções:

a)$f(x)=e^{2x}−e^x$

b)$f(x)=e^x+x$

**Atenção**: O tempo é importante, então escolha, entre usar ferramentas computacionais ou gráficas. Independente da sua escolha você deverá escrever a resposta em uma célula de textos.


```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *

x = symbols('x')

f = exp(2*x) - exp(x)

primeiraDerivada = diff(f, x)
print("Primeira derivada de f(x): ", primeiraDerivada)

segundaDerivada = diff(primeiraDerivada, x)
print("Segunda derivada de f(x): ", segundaDerivada)




f_prime = lambdify(x, primeiraDerivada, 'numpy')

f_double_prime = lambdify(x, segundaDerivada, 'numpy')

x_values = np.linspace(-10, 10, 400)

y_prime_values = f_prime(x_values)

y_double_prime_values = f_double_prime(x_values)

plt.plot(x_values, y_prime_values, label="1ª Derivada de f(x)")

plt.plot(x_values, np.full_like(x_values, y_double_prime_values), 'r--', label="2ª Derivada de f(x)")

plt.xlabel("x")
plt.ylabel("f'(x) / f''(x)")
plt.title("Gráfico das Derivadas de f(x)")
plt.legend()
plt.grid(True)
plt.show()
```

# Questão 4 - Alternativa - Vale até 2 pontos

Em uma fábrica, latas de sopa cilíndricas devem ser fabricadas para conter um volume dado $V$. Não há desperdício de material na corte da parte vertical de cada lata ( o lado do cilíndro), mas cada tampa e fundo, que são círculos de raio $r$, são cortados de um quadrado que mede $2r$ unidades de lado. Assim, o material usado para fabricar cada lata de sopa tem uma área de $A = 2 \pi rh + 8r^2$ unidades quadradas. Encontre a proporção entre a altura e o diâmetro para a lata mais econômica (ou seja, que requer a menor quantidade de material para fabricação).


Precisamos expressar a altura $h$ em termos do raio $r$ usando a fórmula do volume de um cilindro, que é $V = \pi r^2 h$. Isso nos dá $h = \frac{V}{\pi r^2}$.

Substituímos $h$ na fórmula da área $A$ para obter $A$ em termos de $r$ apenas:

$$ A = 2 \pi r \left(\frac{V}{\pi r^2}\right) + 8r^2 = \frac{2V}{r} + 8r^2 $$

Agora, queremos minimizar $A$ em relação a $r$, então vamos encontrar a derivada de $A$ em relação a $r$ e igualar a zero para encontrar o valor de $r$ que minimiza $A$:

$$ \frac{dA}{dr} = -\frac{2V}{r^2} + 16r $$

Igualando a zero e resolvendo para $r$, obtemos:

$$ -\frac{2V}{r^2} + 16r = 0 $$

$$ r^3 = \frac{V}{8} $$

$$ r = \left(\frac{V}{8}\right)^{1/3} $$

Substituímos $r$ na fórmula para $h$ para obter $h$:

$$ h = \frac{V}{\pi r^2} = \frac{V}{\pi \left(\frac{V}{8}\right)^{2/3}} = \left(\frac{8}{\pi}\right)^{1/3} V^{1/3} $$

A proporção entre a altura e o diâmetro (que é $2r$) é então:

$$ \frac{h}{2r} = \frac{\left(\frac{8}{\pi}\right)^{1/3} V^{1/3}}{2\left(\frac{V}{8}\right)^{1/3}} = \frac{1}{2} \left(\frac{8}{\pi}\right)^{1/3} $$

Portanto, a lata mais econômica é aquela em que a altura é aproximadamente $0.5$ vezes o diâmetro.
